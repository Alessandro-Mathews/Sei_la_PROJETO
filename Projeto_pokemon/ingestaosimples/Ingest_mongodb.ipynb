```python
#Importar para fazer o uso:
import pymongo
import pandas as pd
```

```python
#Para carregar o arquivo CSV
df = pd.read_csv('C:/Users/aless/Desktop/Sei_la_PROJETO/Projeto_pokemon/Tratamento_de_dados/Pokemon_alterados.csv')
```

```python
# Conexão com o MongoDB
host = 'localhost'
port = 27017
database_name = 'pokemon_bd'
collection_name = 'pokemon_name_hp_attack_defense_speed'
```

```python
mongo_client = pymongo.MongoClient('localhost', 27017)
pokemon_bd = mongo_client['pokemon_bd']
```

```python
#Obter a coleção do mongodb
mongo_collection = pokemon_bd['pokemon_name_hp_attack_defense_speed']
```

```python
# Verificar as colunas do existentes:
print(df.columns)
```

```python
#Selecionar as colunas 'name', 'hp', 'attack', 'defense' , 'speed'
selected_columns = ['Name', 'HP', 'Attack', 'Defense', 'Speed']
```

```python
#Escolher as colunas selecionadas e muda o nome:
df_selected = df[selected_columns]
df_selected.columns = ['name', 'hp', 'attack', 'defense', 'speed'] 
```

```python
# Converter cada linha do DataFrame em um documento (dicionário) para inserção no mongodb
documents = df_selected.to_dict(orient='records')
```

```python
# Inserir os documentos no mongodb
mongo_collection.insert_many(documents)
```

```python
#Observar a quantidade de pokemons inseridos:
print({len(documents)})
```
